```python
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

penam_core_smiles = "NC1C(=O)N2[C@@H]1SC([C@@H]2C(=O)O)(C)C"
carbamate_smiles   = "O=C(N[C@@H]1C(=O)N2[C@@H]1SC([C@@H]2C(=O)O)(C)C)COc1ccccc1"

m1 = Chem.MolFromSmiles(penam_core_smiles)
m2 = Chem.MolFromSmiles(carbamate_smiles)

if m1 is None or m2 is None:
    raise ValueError("SMILES parse failed. 先检查 SMILES 拼写是否正确。")

Chem.SanitizeMol(m1)
Chem.SanitizeMol(m2)

img1 = Draw.MolToImage(m1, size=(450,300), wedgeBonds=True)
img2 = Draw.MolToImage(m2, size=(450,300), wedgeBonds=True)

print("Penam core SMILES:")
print(penam_core_smiles)
display(img1)

print("\nN-carbamate (phenyl) derivative SMILES:")
print(carbamate_smiles)
display(img2)
```
